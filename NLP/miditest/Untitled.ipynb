```python
from music21 import converter,instrument # or import *
s = converter.parse("./gen_mid/my-sample-song-1.mid")

for el in s.recurse():
    if 'Instrument' in el.classes: # or 'Piano'
        print('hit')
        el.activeSite.replace(el, instrument.TenorSaxophone())

s.write('midi', './gen_mid/out-sample-1.mid')
```
